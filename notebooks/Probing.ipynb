```python
from dataclasses import dataclass

import numpy as np
import pandas as pd


@dataclass
class InputPath:
    _prefix: str = "../input"
    test_images: str = f"{_prefix}/christ-test-imgs.npz"
        
        
@dataclass
class OutputPath:
    _prefix: str = "../output"
    submission: str = f"{_prefix}/submission"


def load_npz(path: str) -> np.array:
    x = np.load(path)["arr_0"]
    return x
```

```python
def create_dummy_sub(pred: int, dry_run: bool) -> None:
    assert pred >= 0 and pred <= 12 and isinstance(pred, int)
    test_images = load_npz(InputPath.test_images)
    df_sub = pd.DataFrame({"id": list(range(len(test_images)))})
    df_sub = df_sub.assign(y=pred)
    print(df_sub.head())
    df_sub = df_sub.astype(int)
    path = f"{OutputPath.submission}/submission_probing_{pred}.csv"
    if dry_run:
        print(f"[dry run]submission will be saved in {path}")
    else:
        df_sub.to_csv(path, index=False)
        print(f"submission is saved in {path}")
```

```python
for i in range(13):
    create_dummy_sub(i, dry_run=False)
```
